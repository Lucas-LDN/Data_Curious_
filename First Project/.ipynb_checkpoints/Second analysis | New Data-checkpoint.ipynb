```python
import numpy as np 
import pandas as pd 
import matplotlib.ticker as ticker 
import matplotlib.pyplot as plt 
import matplotlib.ticker as mtick 
import seaborn as sns 
```

```python
data = pd.read_csv("DH data sheet - Sheet1 (2).csv")

# Add a second column of Impressions, so we can count how many days is was Live.
data["Impressions_2"] = data.Impressions

#Remove commas from Cost and make it a float: 
data["Cost"] = data.Cost.replace('[^\d.]', '', regex = True)                     \
                        .astype(float)

display(data.head(2))

display(data.describe().round(1))

print(f"""\nVariable:                                  Type:\n{data.dtypes}""")
```

## Group into data_ad

```python
#data.columns 
#data.Cost.dtypes.value_counts() 
print(data.dtypes.value_counts()) 
 
display(data.describe().round(2)) 
```

```python
data_ad = data.groupby(["Ad name"])                                           \
              .agg({"Impressions" :  "sum", 
                    "Cost" : "sum", 
                    "Impressions_2" : "count", 
                    "Unique link clicks" : "sum", 
                    "Post shares" : "sum", 
                    "Post comments" : "sum", 
                    "Registration" : "sum", 
                    "Website purchases" : "sum"})                                          \
              .rename(columns = {"Impressions_2" : "Duration"})
 
# Add CPC: 
data_ad["CPC"] = data_ad["Cost"] / data_ad["Unique link clicks"]

# Add CPM: 
data_ad["CPM"] = data_ad["Cost"] / data_ad.Impressions * 1000
    
# Add CTR:
data_ad["CTR"] = data_ad["Unique link clicks"] / data_ad.Impressions 
 
# Add CR: 
data_ad["CR"] = data_ad["Website purchases"] / data_ad["Unique link clicks"]   
 
# Remove NaN values: 
data_ad = data_ad.replace(to_replace = np.nan, value = 0) 
 
# Add a CTR class to data_ad -> Above median average? If yes, it's a 1. If not, it's a 0: 
data_ad["CTR_class"] = pd.cut(data_ad.CTR, 
                              bins = [data_ad.CTR.min() - 1, 
                                      data_ad.CTR.median(), 
                                      data_ad.CTR.max() + 1 ], 
                              labels = ["0", "1"])
     
# Add a CR class to data_ad -> Above median average? If yes, it's a 1. If not, it's a 0: 
data_ad["CR_class"] = pd.cut(data_ad.CR, 
                              bins = [data_ad.CR.min() - 1, 
                                      data_ad.CR.median(), 
                                      data_ad.CR.max() + 1], 
                              labels = ["0", "1"])
 
# Remove inf values: 
data_ad = data_ad.replace(np.inf, 0) 
print("Maximum value in CR is:", data_ad.CR.max().round(2))


display(data_ad.head(3))
display(data_ad.describe().round(2)) 
print(f"""\nVariable:            Type:
{data_ad.dtypes}""")
print(f"\nVariable:              Empty values:\n{data_ad.isnull().sum()}") 
print(data_ad.CTR_class.value_counts())
print(data_ad.CR_class.value_counts())
```

```python
def statistics(variable):
    if variable.dtype == "int64" or variable.dtype == "float64":
        return pd.DataFrame([[variable.name, np.mean(variable), 
                              np.std(variable), np.median(variable), 
                              np.var(variable)]], 
                            columns = ["Variable", "Mean", "Std. Deviation", 
                                       "Median", "Variance"])                      \
                 .set_index("Variable")
    else: 
        return pd.DataFrame(variable.value_counts())
```

```python
def histogram(x): 
    if x.dtype == "int64" or x.dtype == "float64": 
        size_bins = 10
        title = x.name
        color_kde = list(map(float, np.random.rand(3, )))
        color_bar = list(map(float, np.random.rand(3, )))
        sns.distplot(x, bins = size_bins, 
                     kde_kws = {"lw": 1.5, 
                               "alpha" : 0.8, 
                               "color" :  color_kde}, 
                     hist_kws = {"linewidth" : 1.5, 
                                 "edgecolor" : "grey", 
                                 "alpha" : 0.4, 
                                 "color" : color_bar})
    else:
        x = pd.DataFrame(x)
        sns.catplot(x = x.columns[0], 
                    kind = "count", 
                    palette = "spring", 
                    data = x)
```

### Elbow: 

```python
from sklearn.cluster import KMeans 
wcss = [] 

for k in range(1,11): 
    km = KMeans(n_clusters = k, random_state = 0, 
                init = 'k-means++', max_iter = 300, 
                n_init = 10).fit(X)
    wcss.append(km.inertia_)
 
plt.plot(wcss)
plt.title('Elbow Method', size = 14) 
plt.xlabel('Number of clusters', size = 12) 
plt.ylabel('wcss', size = 14) 
plt.show()
```

-> Use 2 Cluster!

```python
kmeans = KMeans(n_clusters = 2, random_state = 0, init = 'k-means++', 
                max_iter = 300, n_init = 10)
kmeans
```

```python
ymeans = kmeans.fit_predict(X)
ymeans 
```

```python
plt.subplots(figsize = (8, 6)) 
 
plt.scatter(transformed_data_ad[:, 0], 
            transformed_data_ad[:, 1], 
            c = ymeans, 
            cmap = plt.cm.get_cmap("Spectral_r", 2)) 

plt.tight_layout() 
plt.show()
```

```python
centroids = pd.DataFrame(kmeans.cluster_centers_, 
                         columns = ["Impressions", "Unique link clicks", 
                                    "Post shares", "Post comments", 
                                    "Duration", "CTR"]) 
 
centroids.index_name = "ClusterID" 
centroids["ClusterID"] = centroids.index 
 
centroids = centroids.reset_index(drop = True) 

centroids["CTR"] = centroids["CTR"] * 100

centroids.round(2)
```

```python
display(data_ad.sample(2))
display(data_ad.describe().round(2))
print(f"\nVariable:                    Empty values:\n{data_ad.isnull().sum()}")
print(f"Variable:                    Type:\n{data_ad.dtypes}")
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
 
X = data_ad.drop(columns = ["Unique link clicks", "CR", "CTR", 
                            "CTR_class"]) 
y = data_ad["Unique link clicks"]
 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, 
                                                    random_state = 1) 
 
model = LinearRegression() 
 
model.fit(X_train, y_train)  
 
model.score(X_test, y_test) 
```

```python
from sklearn.inspection import permutation_importance 
 
result = permutation_importance(model, X, y, 
                                random_state = 5, n_repeats = 100) 
```

```python
result
```
