```python
import numpy as np
import pandas as pd
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
```

```python
data = pd.read_csv("DH data sheet - Sheet1.csv")
```

```python
data.dtypes.value_counts()
```

```python
data["Ad name"].nunique()
```

```python
#convert to date time:
data.Date = pd.to_datetime(data.Date)
data.Date.head(2)
```

```python
#Add Day of the week:
data["day_of_the_week"] = data.Date.dt.weekday

#Change day of the week by adding its name: 
day_names = {0 : "Mon", 1 : "Tue", 2 : "Wed", 3 : "Thu",
            4 : "Fri", 5 : "Sat", 6 : "Sun"}

data["name_of_the_dow"] = data.day_of_the_week.map(day_names)

#Add Day of the month:
data["day_of_the_month"] = data.Date.dt.day

#Add week of the year:
data["week"] = data.Date.dt.week

#Add month:
data["month"] = data.Date.dt.month

data.sample(3)
```

```python
data.name_of_the_dow.value_counts()
```

##  Group by month

```python
data_month = data.groupby(["month"])\
                 ["Impressions", "Unique link clicks", "Signup", "Registration"].sum()       

#adding CTR: 
data_month["CTR"] = data_month["Unique link clicks"] / data_month.Impressions 

#adding CR: 
data_month["CR"] = data_month.Registration / data_month["Unique link clicks"] 

display(data_month.head(3))
display(len(data_month))
```

## Group by week

```python
data_week = data.groupby(["week"])\
                ["Impressions", "Unique link clicks", "Signup", "Registration"].sum()       
data_week.head(3)
```

```python
#adding CTR: 
data_week["CTR"] = data_week["Unique link clicks"] / data_week.Impressions

#adding CR: 
data_week["CR"] = data_week.Registration / data_week["Unique link clicks"]

data_week
```

## Group by ad

```python
data.columns
```

```python
data_ad = data.groupby(["Ad name"])                                 \
              ["Impressions", "Unique link clicks", 
               "Post shares", "Cost", "Cost per website purchase", 
               "Signup", "Registration"].sum()                       \
              .sort_values(["Ad name"],
                           ascending = [False], 
                           inplace = False) 
display(data_ad.head(2))
display(data_ad.shape)
```

```python
data_ad = data.groupby(["Ad name"])                                           \
                     .agg({"Impressions" : "sum", 
                           "Unique link clicks" : "sum", 
                           "Post shares" : "sum", 
                           "Cost" : "sum", 
                           "Cost per website purchase" : "sum", 
                           "Signup" : "sum", 
                           "Registration" : "sum", 
                           "Cost" : "count"}) 
 
data_ad.head() 
```

```python
#Add CTR: 
data_ad["CTR"] = data_ad["Unique link clicks"] / data_ad.Impressions 
 
# Add CR: 
data_ad["CR"] = data_ad.Registration / data_ad["Unique link clicks"] 
 
data_ad.head(2) 
```

```python
# Changing "Cost" into "Duration" of the Ad: 
data_ad.rename(columns = {"Cost": "Duration"}, inplace = True) 
display(data_ad.head(1)) 
display(data_ad.columns) 

# Investigate Duration: 
print("Duration stats: \n", data_ad.Duration.describe())
```

## KMeans Clustering

### Cluster by day:

```python
from sklearn.cluster import KMeans 
```

```python
#Only include numerical values 
clustering_categories = ['Impressions', 'Unique link clicks', 
                         'Post shares', 'Cost per website purchase', 
                         'Signup', 'Registration'] 
 
data_for_clustering = data[clustering_categories] 
display(data_for_clustering.head(3)) 
 
#Check how many Nan values are in there: 
display(data_for_clustering.shape) 
display(data_for_clustering.columns) 
data_for_clustering.isnull().sum() 
```

```python
data_for_clustering = data_for_clustering.replace("NaN", 0)
data_for_clustering.fillna({"NaN" : 0}, inplace = False)

#data_for_clustering.dropna(axis='rows') 
#data_for_clustering.reset_index(drop=True) 

display(data_for_clustering.shape) 
print("""Null values for data_for_clustering:\n""",
      data_for_clustering.isnull().sum()) 
display(data_for_clustering.head()) 
```

```python
intertias = [] 
X = data_for_clustering 
ks = range(1, 10) 

for k in ks: 
    km_elbow = KMeans(n_clusters = k, random_state = 5).fit(X) 
    intertias.append(km_elbow.inertia_)

plt.plot(ks, inertias) 
plt.xlabel(ks)
plt.show()   
```

```python
#Remove the y axis to make this graph more pleasing: 
from sklearn.cluster import KMeans 
from matplotlib.lines import Line2D

intertias = [] 
X = data_ad 
ks = range(1, 10) 

for k in ks: 
    km_elbow = KMeans(n_clusters = k, random_state = 5).fit(X) 
    intertias.append(km_elbow.inertia_) 

fig1 = plt.figure(facecolor = 'white') 
ax1 = plt.axes(frameon = False) 
#Remove the frame on the graph 
ax1.set_frame_on(False) 
ax1.get_xaxis().tick_bottom() 
ax1.axes.get_yaxis().set_visible(False) 
#Remove the ticks on the y axis 

plt.plot(ks, intertias, color = "darkblue") 
plt.xlabel(ks) 
plt.text(9.2, 1,'SSE', color = "darkblue") 

xmin, xmax = ax1.get_xaxis().get_view_interval() 
ymin, ymax = ax1.get_yaxis().get_view_interval() 
ax1.add_artist(Line2D((xmin, xmax), (ymin, ymin), color='darkgrey', linewidth=2)) 

plt.annotate('Optimal number', xy = (3, 2), xytext = (3, 2), 
             arrowprops = dict(facecolor = 'black', shrink = 0.05)) 
plt.show() 
```
