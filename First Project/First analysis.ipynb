# Import, analyse, prepare data

## Import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
data = pd.read_csv("DH data sheet - Sheet1.csv")
```

```python
display(data.head(3))
display(data.shape)
display(data.dtypes)
```

```python
data.dtypes.value_counts()
```

```python
data["Ad name"].nunique()
```

```python
#convert to date time:
data.Date = pd.to_datetime(data.Date)
data.Date.head(2)
```

```python
#Add Day of the week:
data["day_of_the_week"] = data.Date.dt.weekday

#Change day of the week by adding its name: 
day_names = {0 : "Mon", 1 : "Tue", 2 : "Wed", 3 : "Thu",
            4 : "Fri", 5 : "Sat", 6 : "Sun"}

data["name_of_the_dow"] = data.day_of_the_week.map(day_names)

#Add Day of the month:
data["day_of_the_month"] = data.Date.dt.day

#Add week of the year:
data["week"] = data.Date.dt.week

#Add month:
data["month"] = data.Date.dt.month

data.sample(3)
```

```python
data.name_of_the_dow.value_counts()
```

##  Group by month

```python
data_month = data.groupby(["month"])\
                 ["Impressions", "Unique link clicks", "Signup", "Registration"].sum()       

#adding CTR: 
data_month["CTR"] = data_month["Unique link clicks"] / data_month.Impressions 

#adding CR: 
data_month["CR"] = data_month.Registration / data_month["Unique link clicks"] 

data_month 
```

## Group by week

```python
data_week = data.groupby(["week"])\
                ["Impressions", "Unique link clicks", "Signup", "Registration"].sum()       
data_week.head(3)
```

```python
#adding CTR: 
data_week["CTR"] = data_week["Unique link clicks"] / data_week.Impressions

#adding CR: 
data_week["CR"] = data_week.Registration / data_week["Unique link clicks"]

data_week
```

## Group by ad, and week

```python
data_ad_week = data.groupby(["Ad name", "week"])                                 \
                 ["Impressions", "Unique link clicks", "Signup", "Registration"].sum()  \
                 .sort_values(["Ad name", "week"], 
                              ascending = [True, True], 
                              inplace = False)
data_ad_week.head(15)
```

```python
#adding CTR: 
data_ad_week["CTR"] = data_ad_week["Unique link clicks"] / data_ad_week.Impressions

#adding CR: 
data_ad_week["CR"] = data_ad_week.Registration / data_ad_week["Unique link clicks"]

data_ad_week
```

## Group by name of dow

```python
data_grouped_by_dow = data.groupby(["name_of_the_dow"])                                          \
                           ["Impressions", "Unique link clicks", "Signup", "Registration"].sum() \
                          .sort_values(by = "Impressions")

#adding CTR: 
data_grouped_by_weekday["CTR"] = data_grouped_by_weekday["Unique link clicks"] / data_grouped_by_weekday.Impressions

#adding CR: 
data_grouped_by_weekday["CR"] = data_grouped_by_weekday.Registration / data_grouped_by_weekday["Unique link clicks"]

data_grouped_by_weekday
```
